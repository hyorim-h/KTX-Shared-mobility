```python
import pandas as pd
import numpy as np
import requests
```

```python
dtf = pd.read_csv("C:/Users/hyorim/Desktop/연구실/[논문]KTX역사분류/문화시설/철도역인근문화시설_KTX_sample.csv")
```

```python
durations = []
```

```python
for index, row in dtf.iterrows():
    print("index: ", index)

    start_point = str(row['시설경도']) + ',' + str(row['시설위도'])
    end_point = str(row['주요역경도']) + ',' + str(row['주요역위도'])
    X_NCP_APIGW_API_KEY_ID = "000" #애플리케이션 등록 시 발급받은 client id  값
    X_NCP_APIGW_API_KEY = "000" #애플리케이션 등록 시 발급받은 client secret  값

    api_url = "https://naveropenapi.apigw.ntruss.com/map-direction/v1/driving?start=" + start_point + "&goal=" + end_point + "&option=trafast&X-NCP-APIGW-API-KEY-ID=" + X_NCP_APIGW_API_KEY_ID + "&X-NCP-APIGW-API-KEY=" + X_NCP_APIGW_API_KEY
    response = requests.get(api_url)

    # print(response.status_code)
    # print("======")
    # print(response.url)
    # print("======")
    # print(response.text)
    # print("======")
    # print(response.content)
    # print("======")
    # print(response.encoding)
    # print("======")
    # print(response.headers)

    data = response.json()
    summary = data['route']['trafast'][0]["summary"]
    duration = summary['duration'] // 60000
    print(duration)
    durations.append(duration)
```
